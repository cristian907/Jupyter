<h1>Arquitectura de Von Neumann</h1>

<div style="display: flex">
    <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/5/5e/JohnvonNeumann-LosAlamos.gif/250px-JohnvonNeumann-LosAlamos.gif" style="border-radius:10px; max-height:130px; margin-right:10px">
    <p>La Arquitectura de Von Neumann, propuesta por el matemático y físico John von Neumann en 1945, es un diseño conceptual en el cual tanto las instrucciones como los datos son almacenados en un mismo espacio de memoria, utilizando la misma ruta, concepto conocido como "Programa Almacenado". Esta arquitectura permitió la transición de las computadoras de programa fijo, las cuales cumplen una sola funcion específica, a las computadoras de programas almacenados, que se pueden programar para ejecutar muchas tareas diferentes. </p>
</div>

<h2>Componentes Fundamentales</h2>
<p>La arquitectura define la estructura básica de una computadora, organizada en cuatro componentes principales interconectados por buses del sistema (canales de comunicación):</p>
<h4>Unidad Central de Proceso (CPU):</h4>
<p>
  La CPU es el lugar en el que se procesa la información de acuerdo a las instrucciones del programa. La CPU de un sistema informático repite una serie de pasos en los que continuamente accede a memoria para leer la próxima instrucción a ejecutar, realiza lo que ordena la instrucción y vuelta a empezar. Se divide en:

<strong>Unidad Aritmético-Lógica (ALU):</strong> Realiza operaciones matemáticas (suma, resta, etc.) y lógicas (AND, OR, NOT, comparaciones). Normalmente los datos con los que opera, así como los resultados de la operación se encuentran en registros de la CPU.
Las operaciones que es capaz de realizar están definidas por el juego de instrucciones de la CPU. Es posible encontrar ALUs con capacidades orientadas hacia la operación con valores enteros, coma flotante o números imaginarios (por ejemplo en las DSP.)

<strong>Unidad de Control (UC):</strong> La unidad de control gestiona cómo funciona el procesador, enviando señales de control. Decide cómo deben moverse los datos dentro de la computadora, controla las operaciones de entrada y salida y obtiene las instrucciones de la memoria para su ejecución.

<strong>Registros:</strong> Son el tipo de memoria más rapido dentro de la CPU, almacenan temporalmente la información con la que está trabajando el procesador, lo que hace que la ejecución y las operaciones sean más rápidas y eficientes. Es la memoria de trabajo principal de la CPU.
<ul>
    <li>Contador de Programa (PC): Realiza un seguimiento de la direccion de la siguiente instrucción.</li>
    <li>Registro de Instrucciones: Guarda la instrucción actual.</li>
    <li>Registro de Dirección de Memoria: Almacena la dirección de la ubicación de memoria a la que se accede.</li>
    <li>Registro de Datos de Memoria: Retiene temporalmente los datos que se transfieren desde o hacia la memoria.</li>
    <li>Acumulador: Almacena resultados intermedios de operaciones aritméticas y lógicas</li>
</ul>
</p>
<h4>Memoria Principal</h4>
<p>
    La memoria principal tiene por objeto guardar información que es accesible a la CPU. La CPU puede leer y/o escribir datos en las diferentes posiciones de memoria que componen la memoria principal.
La memoria principal tiene menor capacidad que la memoria secundaria (que virtualmente es ilimitada), pero es mucho más rápida. Actualmente la memoria principal se implementa mediante circuitos integrados. La memoria principal de los sistemas informáticos suele estar formada por dos áreas diferenciadas:
    <ul>
        <li>Memoria RAM (Random Access Memory): Memoria de acceso aleatorio (no tiene porqué ser utilizada de manera secuencial) que permite tanto la lectura como la escritura. Habitualmente en los sistemas informáticos se trata de un medio de almacenamiento volátil, de manera que se pierde su contenido al cesar la alimentación.
</li>
        <li>Memoria ROM (Read Only Memory): Memoria de acceso aleatorio que sólo permite la lectura de los datos que almacena. Se trata de un medio de almacenamiento persistente, pues no pierde su contenido cuando cesa la alimentación.</li>
    </ul>
</p>

<h4>Unidades de Entrada/Salida (E/S)</h4>
<p>
    Permiten a la computadora comunicarse con el mundo exterior. Incluyen dispositivos como el teclado, el ratón (entrada), el monitor, la impresora (salida) y el almacenamiento secundario (discos duros, SSD). Se conecta a la CPU y a la memoria mediante un bus de E/S.
</p>
<h4>Buses del Sistema</h4>
<p>
    El bus es un sistema de comunicación que transfiere datos, direcciones y señales de control entre la CPU, la memoria y los dispositivos de E/S. En la arquitectura de Von Neumann, se comparte un único bus tanto para datos como para instrucciones, lo que puede crear un cuello de botella (conocido como cuello de botella de Von Neumann).
</p>

<h2>Proceso de Ejecución</h2>
<p>El funcionamiento de una máquina Von Neumann se basa en un proceso repetitivo llamado Ciclo de Instrucción o Ciclo de Fetch-Decode-Execute (Búsqueda-Decodificación-Ejecución), que la CPU realiza continuamente:

<strong>Búsqueda (Fetch)</strong>: La Unidad de Control lee la siguiente instrucción desde la Memoria Principal, utilizando la dirección almacenada en el Contador de Programa (un registro especial).

<strong>Decodificación (Decode)</strong>: La Unidad de Control interpreta la instrucción que acaba de leer para determinar qué operación se debe realizar (por ejemplo, sumar, cargar un dato, etc.) y qué datos necesita.

<strong>Ejecución (Execute)</strong>: La Unidad de Control coordina a la ALU y a otros componentes para llevar a cabo la operación requerida. Los datos necesarios se mueven desde o hacia la memoria, y la ALU realiza el cálculo.

<strong>Actualización</strong>: El Contador de Programa se incrementa para apuntar a la siguiente instrucción, y el ciclo se repite.</p>

<h2>Cuello de Botella de Von Neumann</h2>
<p>La arquitectura Von Neumann, a pesar de su universalidad y simplicidad, tiene una limitación conocida como el "cuello de botella de Von Neumann".

Dado que la CPU tiene un único bus para acceder tanto a las instrucciones como a los datos, no puede buscarlos simultáneamente.

Esto significa que la CPU a menudo tiene que esperar a que la memoria le entregue la instrucción o el dato (o viceversa), limitando la velocidad máxima a la que la computadora puede operar, ya que la CPU es mucho más rápida que la memoria principal.

El procesamiento de datos también es estrictamente secuencial, ejecutanto instrucciones una por una, incluso con tareas que podrian paralelizarse.

Para mitigar este problema, las computadoras modernas implementan optimizaciones, como las memorias caché, que son memorias ultrarrápidas intermedias que usan una arquitectura de Harvard Modificada (con buses separados para datos e instrucciones a nivel de caché), pero manteniendo el diseño fundamental de Von Neumann en la memoria principal.</p>
