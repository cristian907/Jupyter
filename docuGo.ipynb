# Sintaxis Basica en Go

# Introducción a Go

Go es un lenguaje de programación de tipado estático, compilado y enfocado en la simplicidad. A diferencia de otros lenguajes de scripting, en Go cada celda que queramos ejecutar como un programa completo debe declarar su paquete.

### Declaración de Variables
Go ofrece varias formas de declarar variables:
1. **`var`**: Declaración formal con tipo opcional.
2. **`:=`**: Operador de declaración corta (infiere el tipo automáticamente). Solo se puede usar dentro de funciones.

```go
package main

import "fmt"

func main() {
    // Declaración explícita
    var mensaje string = "Hola desde Go en Jupyter"
    
    // Declaración corta (Inferencia de tipo)
    contador := 42
    precio := 19.99
    esValido := true

    fmt.Printf("Texto: %s\n", mensaje)
    fmt.Printf("Entero: %d, Flotante: %.2f, Booleano: %t\n", contador, precio, esValido)
}
```

# Tipos de Datos en Go

Go es muy explícito con sus tipos. No permite operaciones entre tipos diferentes (por ejemplo, no puedes sumar un `int` y un `float64` sin convertir uno primero).

### Tipos Básicos:
* **bool**: `true` o `false`.
* **string**: Cadena de texto (UTF-8).
* **Enteros**: 
    * Con signo: `int8`, `int16`, `int32`, `int64` y `int` (depende de la arquitectura, usualmente 64 bits).
    * Sin signo (positivos): `uint8` (o `byte`), `uint16`, `uint32`, `uint64`.
* **Flotantes**: `float32`, `float64`.
* **Complex**: `complex64`, `complex128` (para números complejos).

```go
package main

import (
    "fmt"
    "math"
)

func main() {
    // Declaración con tipos explícitos
    var entero int = -10
    var natural uint = 50
    var pi float64 = math.Pi
    var esEstudiante bool = true
    var inicial rune = 'C' // rune es un alias de int32, representa un carácter Unicode

    fmt.Printf("Tipo: %T, Valor: %v\n", entero, entero)
    fmt.Printf("Tipo: %T, Valor: %v\n", natural, natural)
    fmt.Printf("Tipo: %T, Valor: %v\n", pi, pi)
    fmt.Printf("Tipo: %T, Valor: %v\n", esEstudiante, esEstudiante)
    fmt.Printf("Tipo: %T, Valor: %c (Unicode: %d)\n", inicial, inicial, inicial)

    // Conversión de tipos (Casting)
    // En Go es obligatorio ser explícito
    var a int = 10
    var b float64 = 2.5
    resultado := float64(a) + b // Convertimos 'a' a float64 para poder sumar
    fmt.Printf("Resultado conversión: %f\n", resultado)
}
```

# Condicionales (if / else)

Los condicionales nos permiten ejecutar código solo si se cumple una condición. En Go, tienen un par de reglas particulares que los hacen ver más "limpios".

### Características:
1. **Sin paréntesis:** A diferencia de C++ o Java, la condición no va entre paréntesis `()`.
2. **Llaves obligatorias:** Siempre debes usar `{ }`, incluso si solo hay una línea de código.
3. **Declaración corta:** Puedes declarar una variable y evaluarla en la misma línea (muy usado para manejo de errores).

```go
package main

import "fmt"

func main() {
    edad := 18

    // if estándar
    if edad >= 18 {
        fmt.Println("Eres mayor de edad.")
    } else if edad == 17 {
        fmt.Println("Casi eres mayor de edad.")
    } else {
        fmt.Println("Eres menor de edad.")
    }

    // Declaración corta dentro del if (variable limitada solo al bloque if)
    if presupuesto := 500; presupuesto > 200 {
        fmt.Printf("Presupuesto de %d es suficiente.\n", presupuesto)
    }
}
```

# Ciclos (for)

En muchos lenguajes existe `while`, `do-while` y `for`. En Go, **solo existe el `for`**, pero es tan versátil que puede hacer el trabajo de todos los demás.

### Formas de usar el `for`:
* **Tradicional:** Con un contador (inicio; condición; incremento).
* **Como "While":** Solo con una condición (se ejecuta mientras sea verdadera).
* **Infinito:** Sin ninguna condición (útil para servicios que nunca deben parar).
* **Rango:** Para recorrer listas (Slices) o Diccionarios (Maps).

```go
package main

import "fmt"

func main() {
    // 1. Ciclo tradicional (C-style)
    fmt.Print("Contador 1 al 3: ")
    for i := 1; i <= 3; i++ {
        fmt.Printf("%d ", i)
    }
    fmt.Println()

    // 2. Ciclo estilo "While" 
    // (se ejecuta mientras n sea menor a 10)
    n := 1
    for n < 10 {
        n *= 2
    }
    fmt.Printf("Valor final de n: %d\n", n)

    // 3. Ciclo infinito con ruptura (break)
    iter := 0
    for {
        if iter >= 2 {
            break // Sale del ciclo
        }
        fmt.Println("Ciclo infinito controlado")
        iter++
    }
}
```

# Funciones

Una de las características más potentes de Go es que las funciones pueden retornar múltiples valores. Esto se usa comúnmente para retornar un resultado junto con un error (`res, err := funcion()`).

```go
package main

import (
    "fmt"
    "errors"
)

// Función con retorno múltiple
func dividir(a, b float64) (float64, error) {
    if b == 0 {
        return 0, errors.New("no se puede dividir por cero")
    }
    return a / b, nil
}

func main() {
    resultado, err := dividir(10, 2)
    if err != nil {
        fmt.Println("Error:", err)
    } else {
        fmt.Printf("Resultado de la división: %.2f\n", resultado)
    }

    // Ejemplo fallido
    _, errFail := dividir(10, 0)
    fmt.Println("Intento fallido:", errFail)
}
```

# Slices: Listas Dinámicas

Un **Slice** es una secuencia de elementos del mismo tipo que puede crecer o encogerse. Es lo más parecido a una `List` en otros lenguajes o a un array dinámico.

### Conceptos clave:
* **Indexación:** Al igual que en la mayoría de los lenguajes, el primer elemento es el índice `0`.
* **Función `append`:** Se usa para añadir elementos al final del slice.
* **Rango (`range`):** Es la forma más fácil de recorrer los elementos en un bucle.

```go
package main

import "fmt"

func main() {
    // 1. Declarar un slice con valores iniciales
    frutas := []string{"Manzana", "Pera", "Banana"}

    // 2. Añadir elementos con append
    frutas = append(frutas, "Naranja")

    // 3. Acceder por índice
    fmt.Println("La primera fruta es:", frutas[0])

    // 4. Recorrer el slice con un bucle for y range
    fmt.Println("Lista completa:")
    for indice, valor := range frutas {
        fmt.Printf("Índice %d: %s\n", indice, valor)
    }

    // 5. Ver el tamaño
    fmt.Printf("Tienes %d frutas en total.\n", len(frutas))
}
```

# Maps: Diccionarios de Clave-Valor

Un **Map** es una colección desordenada de parejas **clave-valor**. Es ideal para buscar información rápidamente usando una etiqueta (clave) en lugar de un número (índice).

### Conceptos clave:
* **Clave Única:** No puedes tener dos claves iguales.
* **Declaración:** Se define como `map[TipoDeClave]TipoDeValor`.
* **Acceso:** Si buscas una clave que no existe, Go devuelve el "valor por defecto" del tipo (por ejemplo, `0` para números o `""` para strings).

```go
package main

import "fmt"

func main() {
    // 1. Crear un mapa (clave: string, valor: int)
    // Usamos make para inicializarlo vacío
    edades := make(map[string]int)

    // 2. Asignar valores
    edades["Cris"] = 22
    edades["Juan"] = 25
    edades["Jose"] = 30

    // 3. Acceder a un valor
    fmt.Printf("La edad de Cris es: %d\n", edades["Cris"])

    // 4. Eliminar un elemento
    delete(edades, "Jose")

    // 5. Verificar si una clave existe
    // Go devuelve un segundo valor booleano (ok)
    valor, existe := edades["Jose"]
    if !existe {
        fmt.Println("Jose ya no está en el mapa")
    } else {
        fmt.Println("Edad de Jose:", valor)
    }

    // 6. Recorrer el mapa
    fmt.Println("\nLista de edades:")
    for nombre, edad := range edades {
        fmt.Printf("%s tiene %d años\n", nombre, edad)
    }
}
```

# Structs y "Orientación a Objetos" en Go

Go no es un lenguaje orientado a objetos tradicional. No tiene `classes` ni `inheritance` (herencia). En su lugar, usa **Structs** para los datos y **Métodos** para el comportamiento.

### ¿Cómo reemplazan a las clases?
1. **Structs:** Son plantillas que guardan solo **datos** (campos).
2. **Métodos:** Son funciones que se "pegan" a una struct usando un **receptor**.
3. **Encapsulamiento:** En lugar de usar palabras como `public` o `private`, Go usa la capitalización:
   - `Nombre` (con Mayúscula): Es **público** y accesible desde otros paquetes.
   - `edad` (con minúscula): Es **privado** y solo se ve dentro de su paquete.

```go
package main

import "fmt"

// 1. Definimos la estructura (Como si fuera el molde de una clase)
type Usuario struct {
    Nombre string // Público
    Email  string // Público
    edad   int    // Privado (solo accesible en este paquete)
}

// 2. Definimos un MÉTODO para la struct (Comportamiento)
// (u Usuario) es el "receptor". Permite que llamemos a la función como usuario.Saludar()
func (u Usuario) Saludar() {
    fmt.Printf("Hola, mi nombre es %s y mi correo es %s\n", u.Nombre, u.Email)
}

// 3. Método para modificar datos (Receiver por puntero *)
// Usamos '*' porque si no, Go crea una copia y el cambio no se guarda en el original.
func (u *Usuario) CumplirAnios() {
    u.edad++
    fmt.Printf("¡Felicidades %s! Ahora tienes %d años.\n", u.Nombre, u.edad)
}

func main() {
    // Instanciación (Crear un objeto)
    persona := Usuario{
        Nombre: "Cris",
        Email:  "cris@ejemplo.com",
        edad:   22,
    }

    // Llamada a métodos
    persona.Saludar()
    persona.CumplirAnios()
}
```

# Composición
En otros lenguajes, un `Estudiante` heredaría de `Persona`. En Go, usamos **Composición** (meter una struct dentro de otra). Esto se llama "Embedding".

```go
package main

import "fmt"

type Persona struct {
    Nombre string
}

func (p Persona) Presentarse() {
    fmt.Printf("Soy %s. ", p.Nombre)
}

// Estudiante "contiene" a Persona
type Estudiante struct {
    Persona // Esto es Embedding (composición)
    Carrera string
}

func main() {
    est := Estudiante{
        Persona: Persona{Nombre: "Cris"},
        Carrera: "Ingeniería en Computación",
    }

    // El estudiante puede usar los métodos de Persona directamente
    est.Presentarse() 
    fmt.Printf("Estudio %s.\n", est.Carrera)
}
```

# Entrada de Datos (Input)

Para recibir datos del usuario en la consola, Go ofrece principalmente dos caminos:

1. **`fmt.Scanln`**: Ideal para leer palabras sueltas o números. Se detiene cuando encuentra un espacio.
2. **`bufio.NewReader`**: Ideal para leer líneas completas de texto (por ejemplo, una frase con espacios).

### Concepto de Punteros (`&`):
Cuando usamos `Scanln`, pasamos la variable con un **`&`** adelante (ej. `&nombre`). Esto le dice a Go la dirección de memoria de la variable para que pueda "escribir" el valor directamente ahí.

Nota: Jupyter no permite probar el input, codigo solo como ejemplo visual.

```go
package main

import (
    "fmt"
    "bufio"
    "os"
    "strings"
)

func main() {
    // --- OPCIÓN 1: fmt.Scanln (Para una sola palabra) ---
    var nombre string
    fmt.Print("Introduce tu primer nombre: ")
    fmt.Scanln(&nombre) // El & es obligatorio para modificar la variable
    fmt.Printf("Hola, %s!\n", nombre)

    // --- OPCIÓN 2: bufio (Para frases completas con espacios) ---
    // Limpiamos el buffer antes de leer
    reader := bufio.NewReader(os.Stdin)
    fmt.Print("¿Qué estás estudiando?: ")
    
    // Leemos hasta que el usuario presione Enter (\n)
    entrada, _ := reader.ReadString('\n')
    
    // Eliminamos el salto de línea al final para que no ensucie el texto
    entrada = strings.TrimSpace(entrada)
    
    fmt.Printf("¡Genial! No sabía que estudiabas %s.\n", entrada)
}
```
