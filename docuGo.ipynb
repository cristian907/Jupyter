# Sintaxis Basica en Go

## 1. Variables y Tipos de Datos
En Go puedes declarar variables de forma explícita usando la palabra reservada `var`, o usar el operador `:=` para una declaración corta donde Go infiere el tipo de dato automáticamente. 

Los tipos básicos más comunes son `int` (enteros), `float64` (decimales), `string` (texto) y `bool` (booleanos).

```go
func main() {
    // Declaración explícita
    var edad int = 25
    var nombre string = "Alex"

    // Declaración corta (ahora sí es válida porque está dentro de una función)
    altura := 1.75
    esEstudiante := true

    fmt.Printf("Nombre: %s, Edad: %d\n", nombre, edad)
    fmt.Printf("Altura: %.2f, ¿Es estudiante?: %t\n", altura, esEstudiante)
}
```

## 2. Funciones
Las funciones se declaran con la palabra `func`. Debes especificar el tipo de dato de los parámetros que recibe y el tipo de dato de lo que retorna. 

*Nota:* Go permite que una función retorne múltiples valores al mismo tiempo.

```go
// Definimos una función simple
func sumar(a int, b int) int {
    return a + b
}

// Función que retorna múltiples valores
func operaciones(a int, b int) (int, int) {
    suma := a + b
    resta := a - b
    return suma, resta
}

func main() {
    resultadoSuma := sumar(10, 5)
    resSuma, resResta := operaciones(20, 8)

    fmt.Println("Resultado de la suma simple:", resultadoSuma)
    fmt.Println("Múltiples retornos - Suma:", resSuma, "Resta:", resResta)
}
```

## 3. Condicionales (If / Else)
La sintaxis es muy parecida a otros lenguajes, pero con una gran diferencia: **no se usan paréntesis** alrededor de la condición, pero las llaves `{}` son obligatorias siempre.

```go
func main() {
    puntaje := 85

    if puntaje >= 90 {
        fmt.Println("Calificación: A")
    } else if puntaje >= 80 {
        fmt.Println("Calificación: B")
    } else {
        fmt.Println("Calificación: C")
    }
}
```

## 4. Ciclos (Bucles)
En Go, la única palabra clave para hacer ciclos es `for`. No existe `while` ni `do-while`, ya que el `for` de Go es lo suficientemente flexible para cubrir todos esos casos.

```go
func main() {
    // 1. El for tradicional
    fmt.Print("For tradicional: ")
    for i := 0; i < 5; i++ {
        fmt.Printf("%d ", i)
    }
    fmt.Println()

    // 2. El for funcionando como un "while"
    fmt.Print("For como while: ")
    contador := 0
    for contador < 3 {
        fmt.Printf("%d ", contador)
        contador++
    }
    fmt.Println()

    // 3. For iterando sobre un arreglo/slice (con range)
    nombres := []string{"Ana", "Luis", "Carlos"}
    for indice, valor := range nombres {
        fmt.Printf("Índice %d tiene el valor %s\n", indice, valor)
    }
}
```
