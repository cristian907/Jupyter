# Variables y Mutabilidad en Rust

En Rust, la seguridad de la memoria es la prioridad. Por defecto, todas las variables son **inmutables**, lo que significa que una vez que se asigna un valor, este no puede cambiar.

### Conceptos clave:
* **`let`**: Declara una variable inmutable.
* **`let mut`**: Se debe agregar la palabra clave `mut` para permitir que el valor de la variable sea modificado.
* **Shadowing (Sombreado)**: Rust permite declarar una nueva variable con el mismo nombre que una anterior. Esto permite transformar un valor (o cambiar su tipo) manteniendo el nombre de la variable original.

```rust
// Variable inmutable
let nombre = "Cris";

// Variable mutable
let mut edad = 21;
edad = 22; 

// Sombreado (Shadowing) - Cambiamos el tipo de dato en la misma variable
let espacios = "   ";
let espacios = espacios.len(); 

println!("Nombre: {}, Edad: {}, Espacios: {}", nombre, edad, espacios);
```

# Tipos de Datos

Rust es un lenguaje de tipado estático fuerte, lo que significa que debe conocer los tipos de todas las variables en tiempo de compilación.

### Tipos Escalares (Un solo valor):
* **Enteros**: Con signo (`i8`, `i32`, `i64`) y sin signo (`u8`, `u32`, `u64`).
* **Flotantes**: `f32` y `f64` (para números decimales).
* **Booleanos**: `bool` (`true` o `false`).
* **Caracteres**: `char` (representa un valor Unicode de 4 bytes).

### Tipos Compuestos (Múltiples valores):
* **Tuplas**: Agrupan valores de diferentes tipos. Tienen un tamaño fijo.
* **Arrays**: Colecciones de valores del mismo tipo con una longitud fija.

```rust
let entero: i32 = -10;
let flotante: f64 = 3.14159;
let es_estudiante: bool = true;
let inicial: char = 'C';

// Tuplas: Agrupación de tipos heterogéneos
let mi_tupla = (500, 6.4, "Activo");
let (x, y, z) = mi_tupla; // Desestructuración para extraer valores

// Casting (Conversión explícita)
// Rust no permite operaciones entre tipos distintos sin conversión manual
let decimal = 65.9_f32;
let entero_convertido = decimal as u8;

println!("Entero: {}, Flotante: {}, Inicial: {}", entero, flotante, inicial);
println!("Valor extraído de tupla: {}", z);
```

# Funciones y Expresiones

Rust diferencia estrictamente entre sentencias y expresiones:

* **Sentencias (Statements)**: Instrucciones que realizan una acción pero no devuelven un valor. Terminan con punto y coma `;`.
* **Expresiones (Expressions)**: Evalúan un valor. No terminan con punto y coma si se usan como valor de retorno.

### Retorno Implícito:
Si la última línea de una función es una expresión (no tiene punto y coma), Rust la tratará como el valor de retorno de esa función.

```rust
// Definición de función con retorno de tipo i32
fn sumar(a: i32, b: i32) -> i32 {
    a + b  // Sin punto y coma: es el valor que retorna la función
}

// Función sin valor de retorno (realiza una acción)
fn saludar(usuario: &str) {
    println!("Hola, {}! Bienvenido al notebook.", usuario);
}

saludar("Cris");
let resultado = sumar(10, 5);
println!("El resultado es: {}", resultado);
```

# Control de Flujo: Condicionales

Los condicionales `if` permiten bifurcar el código según una condición booleana. En Rust, la condición no requiere paréntesis.

### El if como expresión:
Debido a que `if` es una expresión, puede ser utilizado en el lado derecho de una asignación `let` para determinar el valor de una variable.

```rust
let numero = 7;

if numero < 5 {
    println!("El número es pequeño");
} else if numero == 7 {
    println!("Es el número siete");
} else {
    println!("El número es grande");
}

// Asignación condicionada
let calificacion = if numero >= 5 { "Aprobado" } else { "Reprobado" };
println!("Estado: {}", calificacion);
```

# Control de Flujo: Ciclos

Rust cuenta con tres mecanismos de iteración:

* **`loop`**: Un bucle infinito. Se utiliza `break` para salir y opcionalmente devolver un valor.
* **`while`**: Ejecuta el código mientras la condición sea verdadera.
* **`for`**: Utilizado para iterar sobre elementos de una colección o un rango de números. Es la estructura de ciclo más segura en Rust.

```rust
// Ciclo For sobre un rango (del 1 al 3)
for i in 1..4 {
    println!("Iteración: {}", i);
}

// Ciclo While
let mut n = 3;
while n != 0 {
    println!("Cuenta regresiva: {}", n);
    n -= 1;
}

// Loop con retorno de valor
let mut contador = 0;
let resultado = loop {
    contador += 1;
    if contador == 5 {
        break contador * 10; // Rompe el ciclo y devuelve el valor
    }
};
println!("Resultado del loop: {}", resultado);
```

# Colecciones: Vectores y HashMaps

Las colecciones son estructuras de datos que pueden almacenar múltiples valores en el Heap, permitiendo que su tamaño crezca o disminuya.

* **Vectores (`Vec<T>`)**: Almacenan una lista de valores del mismo tipo uno junto al otro.
* **HashMaps**: Almacenan pares de clave-valor, facilitando la búsqueda de datos mediante una clave única.

```rust
use std::collections::HashMap;

// Vectores: Listas dinámicas
let mut lista = vec![1, 2, 3];
lista.push(4);
println!("Vector completo: {:?}", lista);

// HashMaps: Diccionarios clave-valor
let mut puntuaciones = HashMap::new();
puntuaciones.insert(String::from("Azul"), 10);
puntuaciones.insert(String::from("Rojo"), 50);

println!("Puntos del equipo Azul: {:?}", puntuaciones.get("Azul"));
```

# Estructuras y Métodos

Las estructuras (`struct`) permiten crear tipos de datos personalizados agrupando valores relacionados. Para definir el comportamiento de estas estructuras, se utilizan bloques de implementación (`impl`).

* **Struct**: Define los datos y campos.
* **Impl**: Define los métodos asociados a la estructura.
* **`&self`**: Es el primer parámetro de un método y representa a la instancia de la estructura que está llamando al método.

```rust
struct Computadora {
    marca: String,
    encendida: bool,
}

impl Computadora {
    // Método para mostrar el estado
    fn estado(&self) {
        let mensaje = if self.encendida { "encendida" } else { "apagada" };
        println!("La laptop {} está {}.", self.marca, mensaje);
    }
}

// Creación de una instancia
let mi_pc = Computadora {
    marca: String::from("Arch Linux"),
    encendida: true,
};

mi_pc.estado();
```

# Entrada de Datos (Input)

En Rust, la entrada de datos se gestiona a través del módulo de entrada y salida estándar `std::io`. A diferencia de otros lenguajes, Rust requiere un manejo explícito de posibles errores y el uso de referencias mutables para almacenar la información.

### Conceptos Clave:
* **`String::new()`**: Crea una cadena de texto vacía y crecible donde se guardará la entrada.
* **`stdin().read_line()`**: Lee el flujo de entrada desde el teclado. Requiere una referencia mutable (`&mut`) para poder escribir el texto en la variable.
* **`.expect()`**: Rust obliga a manejar el caso de que la lectura falle (por ejemplo, si no hay un teclado conectado).
* **`.trim()`**: Se utiliza para eliminar el salto de línea (`\n`) que se agrega automáticamente cuando el usuario presiona "Enter".
