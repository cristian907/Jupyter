<h1>Sintaxis basica en Rust</h1>

<h2>Variables y Mutabilidad</h2>
<p>Se utiliza la palabra reservada 'let' para crear variables. En Rust, las variables son inmutables por defecto. Esto significa que una vez que asignas un valor a un nombre, no puedes cambiarlo. Esto ayuda a la concurrencia y a evitar errores de lógica. Para permitir cambios, se usa la palabra clave 'mut'.</p>
<p>Shadowing: Se puede redeclarar una variable con el mismo nombre para cambiar su tipo o valor sin que sea mutable.</p>

```rust
let x = 100; // Inmutable
let mut y = 80; // Mutable

y = y + 10;

// Shadowing: cambiamos de entero a string con el mismo nombre
let x = "Cien"; 

println!("Ejemplo: valor de x = {}, valor de y = {}", x, y);
```

<h2>Tipos de Datos y Seguridad de Tipos</h2>
<p>Rust es de tipado estático fuerte. El compilador debe saber el tamaño de cada dato en tiempo de compilación. Los tipos reflejan el tamaño en bits: u8 (8 bits, 0-255), i32 (32 bits con signo), etc.</p>

```rust
let bit: u8 = 1;              // Entero de 8 bits (un byte)
let pi: f64 = 3.14159;       // Flotante de 64 bits
let es_valido: bool = true;   // Booleano

// Tupla (útil para retornar múltiples valores)
let cpu_status = ("ALU_OK", 0x01); 
let (msg, code) = cpu_status; // Destructuración

println!("Señal: {}, Código: {:x}", msg, code);
```

<h2>Funciones y Expresiones</h2>
<p>En Rust, casi todo es una expresión (devuelve un valor). Una función devuelve el valor de su última línea si esta no tiene punto y coma. Esto hace el código más limpio..</p>

```rust
fn sumar_bits(a: u8, b: u8) -> u8 {
    a + b // Sin ';' es el valor de retorno
}

let resultado = sumar_bits(128, 64);
println!("Resultado suma binaria: {}", resultado);
```

<h2>Condicionales</h2>
<p>Los condicionales if, else if y else permiten ejecutar diferentes bloques de código según se cumpla una condición lógica. En Rust, no necesitas usar paréntesis en la condición y, además, puedes usar el if como una expresión para asignar un valor directamente a una variable, siempre que todos los caminos devuelvan el mismo tipo de dato.</p>

```rust
let numero = 15;

if numero > 20 {
    println!("El número es grande.");
} else if numero > 10 {
    println!("El número es mediano.");
} else {
    println!("El número es pequeño.");
}

// Ejemplo de asignación directa
let mensaje = if numero % 2 == 0 { "Es par" } else { "Es impar" };
println!("Resultado: {}", mensaje);
```

<h2>Ciclos</h2>
<p>Los ciclos permiten repetir tareas de forma eficiente. El ciclo for es el más utilizado porque recorre rangos o listas de forma segura; el ciclo while se ejecuta mientras una condición sea verdadera; y el ciclo loop es un bucle infinito que se detiene manualmente con break. En Rust, el for es especialmente potente porque evita errores al manejar automáticamente los índices.</p>

```rust
// 1. Ciclo FOR: Itera de 1 a 5 (el 6 queda fuera)
println!("Conteo con FOR:");
for i in 1..6 {
    println!("Número: {}", i);
}

// 2. Ciclo WHILE: Se ejecuta mientras la condición sea cierta
let mut n = 3;
println!("\nCuenta regresiva:");
while n > 0 {
    println!("{}", n);
    n -= 1;
}

// 3. Ciclo LOOP: Infinito hasta encontrar un break
let mut contador = 0;
loop {
    contador += 1;
    if contador == 3 {
        println!("\nFin del LOOP en 3.");
        break;
    }
}
```
